```
import random
import math
import matplotlib.pyplot as plt
```

```
def run_ex_pi(trial_no):
  points=[]
  r=10.2
  for _ in range(trial_no):
    x=random.uniform(-1,r)
    y=random.uniform(-1,r)
  points.append([x,y])
  hits_point=[(x,y) for (x,y) in points if math.sqrt(x**2 + y**2) <=r]
  return 4* len(hits_point)/len(points)
 
```

```
if __name__ == '__main__':
  n =int(input("\nEnter the number to generate points: "))
  trial_no=n
```

```
run_ex_pi(trial_no)
```

```
ex_pi=[]
avg_pi_values = []

for _ in range(100):
 ex_pi.append(run_ex_pi(trial_no))
 avg_pi = sum(ex_pi) / len (ex_pi)
 avg_pi_values.append(avg_pi)
print("Experimental Pi Value :",ex_pi)

print("Experimental Average Pi Value :",avg_pi_values)
print("Theoritical Pi Value : %.4f "%math.pi)
avg_pi_errors = [abs(math.pi-pi) for pi in avg_pi_values]
print("Experimental Average Pi Errors :",avg_pi_errors)

 
```

```
plt.xlabel("Trial_no")
plt.ylabel("Error_value")
plt.plot(avg_pi_errors)
```
